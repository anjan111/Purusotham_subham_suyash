```python
# import data science lib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_file = np.genfromtxt('data_file.txt',delimiter = ',')
print data_file
```

```python
time = data_file[:,][:,0]
print time
```

```python
time = time-time[0]
print time
```

```python
sensor = data_file[:,][:,1:5]
print sensor
```

```python
avg_all =np.mean(sensor,axis = 1)
print avg_all
```

```python
avg_ind =np.mean(sensor,axis = 0)
print avg_ind
```

```python
print "time shape : ",time.shape, " time dim : ",time.ndim
print "sensor shape : ",sensor.shape, " sensor dim : ",sensor.ndim
print "avg_all shape:",avg_all.shape," avg_all dim : ",avg_all.ndim
print "avg_ind shape : ",avg_ind.shape," avg_ind dim: ",avg_ind.ndim
```

```python
time = time.reshape(1200,1)
avg_all=avg_all.reshape(-1,1)
avg_ind=avg_ind.reshape(-1,1)
```

```python
print "time shape : ",time.shape, " time dim : ",time.ndim
print "sensor shape : ",sensor.shape, " sensor dim : ",sensor.ndim
print "avg_all shape:",avg_all.shape," avg_all dim : ",avg_all.ndim
print "avg_ind shape : ",avg_ind.shape," avg_ind dim: ",avg_ind.ndim
```

```python
export_data = np.concatenate((time,sensor,avg_all),axis=1)

np.savetxt('export_data_numpy.txt',export_data,delimiter=",")
```

```python
plt.figure(1)
plt.plot(time,sensor[:,][:,0],'r*')
plt.plot(time,avg_all,'b.')
plt.legend(['sensor 1','avg'])
plt.xlabel('time in sec')
plt.ylabel('sensor1')
plt.savefig("my_senso1.png")
plt.show()
```
